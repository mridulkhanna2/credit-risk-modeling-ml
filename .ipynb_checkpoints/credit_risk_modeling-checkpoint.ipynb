```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
```

## Group Component 1

```python
df = pd.read_csv('credit_score_classification.csv')
df
```

```python
#Removing all numeric values and characters 

df = df[~df['Extra_Column_1'].str.contains(r'\d|\?')]
df = df[~df['Extra_Column_2'].str.contains(r'\d|\?')]
df = df[~df['Extra_Column_3'].str.contains(r'\d|\?')]
```

```python
#Concatenating the columns 'Extra_Column_1', Extra_Column_2', 'Extra_Column_3' to Loan_type 
df['Loan_Types'] = df['Loan_Types'] + ', ' + df['Extra_Column_1'] + ', ' + df['Extra_Column_2'] + ', ' + df['Extra_Column_3']
df.drop(['Extra_Column_1', 'Extra_Column_2', 'Extra_Column_3'], axis=1, inplace=True)
```

```python
#Keeping only int age values between 14 to 100 years
try:
    df['Customer_Age'] = pd.to_numeric(df['Customer_Age'])
except ValueError:
    df['Customer_Age'] = pd.to_numeric(df['Customer_Age'], errors='coerce')
df = df[(df['Customer_Age'] >= 14) & (df['Customer_Age'] <= 100)]
df = df.dropna(subset=['Customer_Age'])
df['Customer_Age'] = df['Customer_Age'].astype(int)
```

```python
# Step 1: Replace '?' and other non-numeric values with NaN.
df['Outstanding_Debt'] = pd.to_numeric(df['Outstanding_Debt'], errors='coerce')

# Step 2: Check for NaN values and decide how to handle them.
# Option 1: Fill NaN values with the mean of the column.
df['Outstanding_Debt'].fillna(df['Outstanding_Debt'].mean(), inplace=True)

# Option 2: Alternatively, you can drop rows with NaN values if preferred.
# df.dropna(subset=['Outstanding_Debt'], inplace=True)

# Step 3: Convert the column to float (it should already be float from pd.to_numeric).
df['Outstanding_Debt'] = df['Outstanding_Debt'].astype('float')

# Verify the conversion.
print(df['Outstanding_Debt'].head())
```

```python
df.head()
```

```python
#Removing empty lines from the column 
df['Customer_Occupation'] = df['Customer_Occupation'].replace('_______', 'NA')

#Replacing characters from 'Customer_SSN" and 'Monthly_salary'
#Removing '$' from all rows and changing name of the column 'Monthly_salary'
df['Customer_SSN'] = df['Customer_SSN'].replace('#F\%$D@*&8', 'NA')

df['Monthly_Salary'] = df['Monthly_Salary'].replace('?', 'NA')
df["Monthly_Salary"] = df["Monthly_Salary"].str.replace("$", "", regex=False)
df.rename(columns={"Monthly_Salary": "Monthly_Salary (in $)"}, inplace=True)
```

```python
#Removing unwanted characters from the data
df['Loan_Types'] = df['Loan_Types'].str.replace("'", '', regex=False)
```

```python
#Replacing noise data with NA
df['Payment_Delay_Days'] = pd.to_numeric(df['Payment_Delay_Days'], errors='coerce').fillna('NA')
df['Delayed_Payments'] = pd.to_numeric(df['Delayed_Payments'], errors='coerce').fillna('NA')
df['Credit_Limit_Changes'] = pd.to_numeric(df['Credit_Limit_Changes'], errors='coerce').fillna('NA')
df['Credit_Inquiries'] = pd.to_numeric(df['Credit_Inquiries'], errors='coerce').fillna('NA')
```

```python
df
```

```python
#Keeping only 3 categories for 'Credit_Mix' and making the blanks and noise as not given 
df['Credit_Mix'] = df['Credit_Mix'].apply(lambda x: x if x in ['Good', 'Standard', 'Bad'] else 'Not Given')
```

```python
#Replacing non-numeric values with NA 
df['Credit_Utilization'] = pd.to_numeric(df['Credit_Utilization'], errors='coerce')
df['Credit_Utilization'] = df['Credit_Utilization'].fillna('NA')
```

```python
df['Credit_History_Age'] = df['Credit_History_Age'].apply(lambda x: 'NA' if isinstance(x, (int, float)) or x in ['Standard', 'Bad','_','?'] else x)
```

```python
pattern = r'(\d+ Years and \d+ Months)'

# Filling non-matches with spaces
df['Credit_History_Age'] = df['Credit_History_Age'].astype(str).str.extract(pattern, expand=False).fillna(' ')
```

```python
df['Min_Amount_Payment'] = df['Min_Amount_Payment'].apply(lambda x: 'NG' if (x not in ['No', 'Yes']) else x)
```

```python
#Keeping only numeric values and replacing other values with NM
df['Monthly_EMI'] = pd.to_numeric(df['Monthly_EMI'], errors='coerce')
df['Monthly_EMI'] = df['Monthly_EMI'].fillna('NA')

df['Monthly_Investment'] = pd.to_numeric(df['Monthly_Investment'], errors='coerce')
df['Monthly_Investment'] = df['Monthly_Investment'].fillna('NA')

df['Monthly_Balance'] = pd.to_numeric(df['Monthly_Balance'], errors='coerce')
df['Monthly_Balance'] = df['Monthly_Balance'].fillna('NA')
```

```python
allowed_categories = ["High_spent_Small_value_payments", "Low_spent_Large_value_payments", "Low_spent_Medium_value_payments", "Low_spent_Small_value_payments","High_spent_Medium_value_payments","High_spent_Large_value_payments"]

# Replacing values not in allowed_categories with spaces
df['Payment_Behavior'] = df['Payment_Behavior'].apply(lambda x: x if x in allowed_categories else 'NA')
```

```python
allowed_categories = [ "Good", "Poor", "Standard", "Bad"]

# Replacing values not in allowed_categories with spaces
df['Credit_Score'] = df['Credit_Score'].apply(lambda x: x if x in allowed_categories else 'NA')
```

```python
#Dropping columns
df = df.drop(['Extra_Column_5', 'Extra_Column_6', 'Extra_Column_7', 'Extra_Column_8', 'Extra_Column_4','Unnamed: 0'], axis=1)
```

```python
# Rounding off columns to 2 decimals
round_off = [ 'Monthly_Balance','Monthly_EMI', 'Annual_Income', 'Outstanding_Debt', 'Monthly_Investment','Monthly_Salary (in $)']
df[round_off] = df[round_off].apply(lambda x: pd.to_numeric(x, errors='coerce').round(2))
```

```python
# Removing outliers from interest rate ( interest_rate <= 50)
df = df[df['Interest_Rate'] <= 50]
df.rename(columns={'Interest_Rate': 'Interest_Rate (in %)'}, inplace=True)
```

```python
# Removing negative values from Credit_Inquiries and Credit_Utilization : as not possible

df['Credit_Inquiries'] = pd.to_numeric(df['Credit_Inquiries'], errors='coerce')
df['Credit_Utilization'] = pd.to_numeric(df['Credit_Utilization'], errors='coerce')

df = df[(df['Credit_Inquiries'] >= 0) & (df['Credit_Utilization'] >= 0)]
df = df.dropna(subset=['Credit_Inquiries', 'Credit_Utilization'])
```

```python
# Standardizing unknown values
df['Credit_Mix'].replace('Not Given', np.nan, inplace=True)
df['Min_Amount_Payment'].replace('NG', np.nan, inplace=True)
```

```python
#handling missing values
df = df.drop(columns=['Monthly_Investment']) #irrelevant to our research and has lots of missing values
missing_occupations = df['Customer_Occupation'].isna().sum()


df['Customer_Occupation'].replace('NA', np.nan, inplace=True)
df = df.dropna(subset=['Customer_Occupation'])
# Checking unique values
unique_occu = df['Customer_Occupation'].unique()
print(f"Number of missing values in 'Customer_Occupation': {missing_occupations}")
print("Unique Customer Occupations:", unique_occu)
```

```python
#Checking outliers in the dataset
df.describe()
```

```python
df['Annual_Income'] = pd.to_numeric(df['Annual_Income'], errors='coerce')
df['Outstanding_Debt'] = pd.to_numeric(df['Outstanding_Debt'], errors='coerce')
df['Interest_Rate (in %)'] = pd.to_numeric(df['Interest_Rate (in %)'], errors='coerce')
```

### 1. UNIKEY: MKHA0689 - XGBOOST

```python
# Importing all required libraries
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score, f1_score, roc_auc_score, classification_report, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import RandomizedSearchCV
```

##### Step 1 : Feature Scaling
Scaling numeric columns as the range of distribution is very different for numeric columns

```python
# Defining all the numeric columns to use for feature scaling
num_columns=['Loans', 'Annual_Income',  'Credit_Utilization', 'Monthly_Balance','Outstanding_Debt', 'Interest_Rate (in %)', 'Income_to_Debt_Ratio',
                'Monthly_Salary (in $)', 'Credit_Inquiries', 'Monthly_EMI', 'Credit_Cards']
scaler=StandardScaler()
train_data[num_columns] = scaler.fit_transform(train_data[num_columns])
test_data[num_columns] = scaler.transform(test_data[num_columns])
val_data[num_columns] = scaler.transform(val_data[num_columns])
```

##### Step 2: Feature Engineering
Using feature engineering to create new relation between variable  and modify it

```python
# Adding the column Debt-Income as it directly relates with the research question
train_data['Debt_to_Income_Ratio']=train_data['Outstanding_Debt'] / train_data['Annual_Income']
val_data['Debt_to_Income_Ratio']=val_data['Outstanding_Debt'] / val_data['Annual_Income']
test_data['Debt_to_Income_Ratio']=test_data['Outstanding_Debt'] / test_data['Annual_Income']

# To relate the spending with income
train_data['Income_Credit']=train_data['Credit_Utilization']*train_data['Annual_Income']
val_data['Income_Credit']=val_data['Credit_Utilization']*val_data['Annual_Income']
test_data['Income_Credit']=test_data['Credit_Utilization']*test_data['Annual_Income']

# Applying log tranform for income because its skewed
train_data['Log_Annual_Income'] = np.log1p(train_data['Annual_Income'])
val_data['Log_Annual_Income'] = np.log1p(val_data['Annual_Income'])
test_data['Log_Annual_Income'] = np.log1p(test_data['Annual_Income'])
```

##### Step 3: Feature-Target Split

```python
# Dividing the data into target variable and features

# Removing the target column from this to feed the model
train_x = train_data.drop(columns=['Customer_Occupation'])
test_x = test_data.drop(columns=['Customer_Occupation'])
validation_x = val_data.drop(columns=['Customer_Occupation'])

#Creating a new variable for target column
train_y = train_data['Customer_Occupation']
validation_y = val_data['Customer_Occupation']
test_y = test_data['Customer_Occupation']
```

```python
# Adding the column Debt-Income as it directly relates with the research question
train_data['Debt_to_Income_Ratio']=train_data['Outstanding_Debt'] / train_data['Annual_Income']
val_data['Debt_to_Income_Ratio']=val_data['Outstanding_Debt'] / val_data['Annual_Income']
test_data['Debt_to_Income_Ratio']=test_data['Outstanding_Debt'] / test_data['Annual_Income']

# Interaction feature between Annual_Income and Credit_Utilization
train_data['Income_Credit_Interaction'] = train_data['Annual_Income'] * train_data['Credit_Utilization']
val_data['Income_Credit_Interaction'] = val_data['Annual_Income'] * val_data['Credit_Utilization']
test_data['Income_Credit_Interaction'] = test_data['Annual_Income'] * test_data['Credit_Utilization']

# Log transformation of Annual_Income (helps with skewed distributions)
train_data['Log_Annual_Income'] = np.log1p(train_data['Annual_Income'])
val_data['Log_Annual_Income'] = np.log1p(val_data['Annual_Income'])
test_data['Log_Annual_Income'] = np.log1p(test_data['Annual_Income'])
```

##### Step 4: Label Encoding
Encoding the target variable i.e Customer_Occupation is necessary as XGBoost, as XGBoost works with numerical data

```python
label_encoder = LabelEncoder()

train_y_encoded = label_encoder.fit_transform(train_y)
validation_y_encoded = label_encoder.transform(validation_y)
test_y_encoded = label_encoder.transform(test_y)
```

```python
#Printing encoded values
label_mapping = dict(zip(label_encoder.classes_, label_encoder.transform(label_encoder.classes_)))
label_df = pd.DataFrame(list(label_mapping.items()), columns=['Occupation', 'Encoded_value'])
print(label_df.to_string(index=False, justify='center'))
```

```python
# Drop irrelevant columns from the dataset
#These are obtained after evaluating the feature importance code below  
irrelevant_columns = ['Record_ID', 'Customer_ID', 'Record_Month', 'Customer_SSN', 
                      'Bank_Accounts', 'Payment_Delay_Days', 'Delayed_Payments', 
                      'Credit_Limit_Changes', 'Credit_History_Age', 
                      'Min_Amount_Payment', 'Payment_Behavior']

train_x = train_x.drop(columns=irrelevant_columns)
test_x = test_x.drop(columns=irrelevant_columns)
validation_x = validation_x.drop(columns=irrelevant_columns)
```

```python
#Defining the feature columns to feed  
features = ['Customer_Age',  'Debt_to_Income_Ratio''Interest_Rate (in %)', 'Credit_Utilization','Outstanding_Debt', 'Income_Credit', 'Log_Annual_Income']

X_train = train_data[features]
X_val = val_data[features]
X_test = test_data[features]
```

## 1. Model Development (without optimization)

##### Training the model 

```python
xgboost_model = XGBClassifier(random_state=29, use_label_encoder=False, eval_metric='mlogloss')
xgboost_model.fit(X_train, train_y_encoded)
```

##### Evaluation on Validation Dataset

```python
# Evaluating on validation set
y_val_pred = xgboost_model.predict(X_val)
# Using predict_proba() for calculating probabilities of AUC-ROC
y_val_proba = xgboost_model.predict_proba(X_val) 

# Calculating accuracy
val_accuracy = accuracy_score(validation_y_encoded, y_val_pred)
#Calculating F1 Score
val_f1 = f1_score(validation_y_encoded, y_val_pred, average='macro')
# Calculating AUC-ROC score
val_auc_roc = roc_auc_score(validation_y_encoded, y_val_proba, multi_class='ovr', average='macro')

print('VALIDATION DATASET')
print(f'Accuracy: {val_accuracy*100:.2f}%')
print(f'F1-Score (Macro): {val_f1*100:.4f}')
print(f'AUC-ROC (Macro): {val_auc_roc:.4f}')

# Generating classification Report
print('Classification Report for Validation dataset:')
print(classification_report(validation_y_encoded, y_val_pred, target_names=label_encoder.classes_))

#Confusion Matrix
val_cm = confusion_matrix(validation_y_encoded, y_val_pred)

cmap = sns.light_palette('#E6E6FA', as_cmap=True, reverse=True) 
plt.figure(figsize=(8, 6))  # Adjust the figure size
sns.heatmap(val_cm,annot=True,cmap=cmap,fmt='d', xticklabels=label_encoder.classes_,yticklabels=label_encoder.classes_)
plt.title('Confusion Matrix - Validation') 
plt.xlabel('Predicted Value') 
plt.xticks(rotation=45, ha='right')
plt.ylabel('True Value') 
plt.tight_layout() 
plt.show()
```

##### Evaluation on Test Dataset

```python
#Evaluating on test dataset
y_test_pred = xgboost_model.predict(X_test)
# Using predict_proba() for calculating probabilities of AUC-ROC
y_test_proba = xgboost_model.predict_proba(X_test)

# Calculating Accuracy
test_accuracy = accuracy_score(test_y_encoded, y_test_pred)
# Calculating F1 Score (macro)
test_f1 = f1_score(test_y_encoded, y_test_pred, average='macro')
# Calculating AUC-ROC
test_auc_roc = roc_auc_score(test_y_encoded, y_test_proba, multi_class='ovr', average='macro')

print('TEST DATASET')
print(f'Accuracy: {test_accuracy*100:.2f}%')
print(f'F1-Score (Macro): {test_f1:.4f}')
print(f'Test AUC-ROC (Macro): {test_auc_roc:.4f}')

# Generating Classification Report
print('Classification Report - Test:')
print(classification_report(test_y_encoded, y_test_pred, target_names=label_encoder.classes_))

# Confusion Matrix 
test_cm = confusion_matrix(test_y_encoded, y_test_pred)

cmap = sns.light_palette('#E6E6FA', as_cmap=True, reverse=True)  # Pink to purple gradient
plt.figure(figsize=(8, 6)) 
sns.heatmap(test_cm,annot=True,cmap=cmap,fmt='d', xticklabels=label_encoder.classes_,yticklabels=label_encoder.classes_)
plt.title('Confusion Matrix - Test') 
plt.xlabel('Predicted Value') 
plt.xticks(rotation=45, ha='right')
plt.ylabel('True Value') 
plt.tight_layout() 
plt.show()
```

## Unikey: AKUM0835  - Decision Tree Model 

Function DecisionTree(data):
    If all data points in the set belong to the same class:
        Return a leaf node with that class
    Else:
        For each feature:
            Compute Gini Impurity/Information Gain for potential splits
        Select the feature and split point with the best Gini/Information Gain
        Split the dataset into subsets based on the selected feature
        Create a decision node with the feature and split point
        For each subset:
            If the stopping condition is met (pure subset, max depth, etc.):
                Return a leaf node with the majority class
            Else:
                Recursively call DecisionTree() on the subset

Return the root node of the tree

```python
# Required imports
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import label_binarize
from sklearn.metrics import roc_curve, auc
import matplotlib.pyplot as plt

# Load your dataset (replace with your dataset path or variable)
df = pd.read_csv('Assignment2_cleaned_data.csv')  # Replace with the actual path or dataframe

# Specify the features and target
features = ['Customer_Age', 'Annual_Income', 'Interest_Rate (in %)', 'Credit_Utilization', 'Outstanding_Debt']
target = 'Customer_Occupation'

# Split the data into training and test sets
train_data, test_data = train_test_split(df, test_size=0.4, stratify=df[target], random_state=29)

# Prepare the data
X_train = train_data[features]
y_train = train_data[target]
X_test = test_data[features]
y_test = test_data[target]

# Feature scaling
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Refit the Decision Tree model using the optimal parameters (from grid search)
dt_classifier_optimized = DecisionTreeClassifier(
    max_depth=None,
    max_features=None,
    min_samples_leaf=1,
    min_samples_split=2,
    random_state=42
)
dt_classifier_optimized.fit(X_train_scaled, y_train)

# Binarize the output labels for multiclass ROC calculation
y_test_binarized = label_binarize(y_test, classes=df[target].unique())
n_classes = y_test_binarized.shape[1]

# Predict probabilities for the test set
y_test_prob = dt_classifier_optimized.predict_proba(X_test_scaled)

# Compute ROC curve and ROC area for each class
fpr = dict()
tpr = dict()
roc_auc = dict()

for i in range(n_classes):
    fpr[i], tpr[i], _ = roc_curve(y_test_binarized[:, i], y_test_prob[:, i])
    roc_auc[i] = auc(fpr[i], tpr[i])

# Plot ROC curves
plt.figure(figsize=(12, 8))
colors = ['aqua', 'darkorange', 'cornflowerblue', 'green', 'red', 'purple', 'brown', 'pink', 'grey', 'yellow']

for i in range(n_classes):
    plt.plot(fpr[i], tpr[i], color=colors[i % len(colors)], lw=2,
             label=f'ROC curve (area = {roc_auc[i]:0.2f}) for class {df[target].unique()[i]}')

plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curves for Each Class')
plt.legend(loc="lower right")
plt.show()
```
